```python
# %%
import numpy as np
import matplotlib.pyplot as plt

# %%
# Create 1D data
x = np.linspace(0, 10, 100)  # 0 to 10, 100 points
y = np.sin(x)                 # sine function

# %%
# Plot the 1D graph
plt.plot(x, y, label="sin(x)", color="blue", marker="")
plt.title("1D Graph of sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
