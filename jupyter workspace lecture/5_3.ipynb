# IMP CONCEPT

```python
import pandas as pd
a = pd.read_excel("titanic.xls", sheet_name="empty")
a  # here a holds some empty cells which are defined as NaN
```

Counting the cells which are empty

```python
# isnull().sum()
a.isnull().sum()        # it will print 6 rows being empty in column 'key'
```

Removing all the rows and columns that holds empty cells

```python
x1 = a.copy()        # to avoid making changes into the original sheet we copied a into x1
x2 = a.copy()        # to avoid making changes into the original sheet we copied a into x2

# dropna -- used to remove the row or column that holds empty cells

# removing rows that holds empty cells
x1.dropna(inplace=True, axis=0)
x1
```

```python
# removing columns that holds empty cells
x2.dropna(inplace=True, axis=1)
x2
```

Removing the duplicate rows and columns

```python
x3 = a.copy()
x3.drop_duplicates(inplace=True)        # this is used to remove all the duplicate rows only, not columns 
```

Replacing the empty cells a value (int/string)

```python
a
```

```python
x4 = a.copy()
x4 = x4.fillna(25)
x4
```

```python
x5 = a.copy()
x5.fillna("AGE NOT AVAILABLE", inplace=True)
x5
```

```python
x6 = a.copy()
x6 = x6.fillna(a["keys"].median())      # here what i did is, i took all the elements of "keys" and found median of all those nums and stored it in the empty cells of x6
x6
```

```python
x7 = a.copy()
x7 = x7.fillna(pow(2, 5))
x7
```
