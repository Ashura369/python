```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# --------------------------------------------------------------------
# for reviewing the learning
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
# --------------------------------------------------------------------


```

```python

df = pd.read_excel("employee_performance_data.xlsx")
df
```

```python
x = df[['Experience_Years', 'Hours_Worked_Per_Week', 'Projects_Handled', 'Training_Score']].values
y = df['High_Performer'].values

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3)

reg = LogisticRegression()
reg.fit(xTrain, yTrain)

yPrediction = reg.predict(xTest)

data = pd.DataFrame({
    'Experience_Years': xTest[:, 0],
    'Hours_Worked_Per_Week': xTest[:, 1],
    'Projects_Handled': xTest[:, 2],
    'Training_Score': xTest[:, 3],
    'ACTUAL OUTPUT': yTest,
    'PREDICTED OUTPUT': yPrediction
})

data
```

```python
a = np.array([[10, 20, 5, 40]])
ans = reg.predict(a)
ans
```

```python
# shows the accuracy, the higher the better
print(accuracy_score(yTest, yPrediction))

# How well the model explains variance in the target (regression fit)
print(r2_score(yTest, yPrediction))

# Average absolute difference between actual and predicted values
print(mean_absolute_error(yTest, yPrediction))

# Average squared difference between actual and predicted values, penalizing larger errors
print(mean_squared_error(yTest, yPrediction))
```
