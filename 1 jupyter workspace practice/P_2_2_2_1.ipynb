```python
import numpy as np

# making 2D array
a = np.arange(10, 101, 10).reshape(2,5)
d = np.random.randint(1,10, (5,5))
print(a,'\n')
print(d,'\n')

# making 3D array
b = np.arange(1, 51).reshape(2,5,5)
b1 = np.arange(1, 51).reshape(1,25,2)
c = np.ones((2,4,3))
d1 = np.random.randint(1,10, (3,5,5))
print(b,'\n')
print(b1,'\n')
print(c,'\n')
print('Making random integers : \n',d1,'\n')






```

```python
# using tile
n = np.arange(0, 40, 10)
f = np.tile(n, (4,2))
print(f"f -- \n {f} \n")

f1 = np.tile(n, (4,2))
print(f"f1 -- \n {f1} \n")

print(f"f * f1 : \n {f*f1} \n")

# using Transpose
f1_Transpose = f1.T
print(f"f1.T -- \n {f1_Transpose} \n")

```
