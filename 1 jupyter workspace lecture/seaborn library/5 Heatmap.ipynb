```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = np.linspace(1,10,20).reshape(4,5) # 4 row 5 columns, the heat map will be of 4 row and 5 columns
# nums should be btwn 1 to 10 and the total numbers should be 20, hence the reshape is 4*5 = 20

var
```

```python
sns.heatmap(var)
```

```python
data =  sns.load_dataset('anagrams')
data.head(5)        # you can only make heatmap of the numerical datas
```

```python
# dropping the unnecessary columns
x = data.drop(columns=['attnr'], axis=1).copy()
x.head(5)
```

```python
sns.heatmap(x)
plt.show()
```

```python
# using annot to show nums inside the colors
sns.heatmap(x, vmin=0, vmax=20, annot=True)
plt.show()
```

```python
# setting line width and line color
sns.heatmap(x, vmin=0, vmax=20, annot=True, linewidth=0.75, linecolor='y')
plt.show()
```

```python
# removing colorbar
sns.heatmap(x, vmin=0, vmax=20, cbar=False)
plt.show()
```

```python
sns.heatmap(x.head(10))     # heatmap of only 10 rows
plt.show()
```

```python
sns.heatmap(x, vmin=0, vmax=20)
# vmin → sets the minimum value for the color scale (the lowest color shade).
# vmax → sets the maximum value for the color scale (the darkest or most intense color).

plt.show()
```

```python
# changing the color map in the graph
sns.heatmap(x, vmin=0, vmax=20, cmap='PuOr')
plt.show()
```
