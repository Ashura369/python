```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset('penguins')
var
```

```python
sns.barplot(var, x='island', y='bill_length_mm')
plt.show()
```

```python
sns.barplot(var, x='island', y='bill_length_mm', hue='sex')
plt.show()
```

```python
sns.barplot(var, x='island', y='bill_length_mm', hue='sex', hue_order=['Female', 'Male'])
plt.show()
```

```python
order_1 = ['A', 'B', 'C']
# will change the name of the columns
sns.barplot(var, x='island', y='bill_length_mm', hue='sex', order = order_1)
plt.show()
```

```python
# changing the orientation of the  graph
# to use this function both of the columns must need to be numerical, if any one of them is textual it will not work
sns.barplot(var, x='bill_depth_mm', y='bill_length_mm', hue='sex', orient='h')
plt.show()
```
