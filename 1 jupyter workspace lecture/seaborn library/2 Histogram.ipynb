```python
var = sns.load_dataset('penguins')
var
```

```python
sns.displot(var, x='bill_length_mm')
plt.show()
```

```python
sns.displot(var, x='bill_length_mm', bins=60)
plt.show()
```

```python
sns.displot(var, x='bill_length_mm', bins=60, kde=True)     # kde - kernal density estimate
plt.show()
```

```python
sns.displot(var, x='bill_length_mm', bins=60, kde=True, rug=True)   
plt.show()

# A rug plot gives you a quick visual of the actual data points in your distribution.
# It helps you see where the data is dense or sparse â€” i.e., where values actually occur.
```

```python
sns.displot(var, x='bill_length_mm', bins=60, kde=True, rug=True, log_scale=True)   
plt.show()


"""    
It applies a logarithmic scale to one or both axes, instead of the regular (linear) scale.

Sometimes your data has values spread across many orders of magnitude (for example, 1, 10, 1000, 100000).
In such cases, a linear axis squishes small values and stretches large ones â€” making patterns hard to see.
Using a log scale spreads out the data more evenly.

"""
```
