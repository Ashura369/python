```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset('tips')
var
```

```python
sns.violinplot(var, x='day', y='total_bill', hue='time')
plt.show()

# a powerful way to visualize the distribution of numerical data across different categories
# it shows the frequency of the data
# the max number of bill was btwn 10 to 20

# â†’ wider = more data points,
# â†’ thinner = fewer data points.
```

```python
sns.violinplot(var, x='time', y='total_bill', hue='time', order=['Dinner', 'Lunch'])    # interchanging the two columns
plt.show()
```

```python
sns.violinplot(var, x='time', y='total_bill', hue='time', split=True)    # interchanging the two columns
plt.show()
```
