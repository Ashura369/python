# Learning Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Difference btwn matplotlib and seaborn  

```python
a = [i for i in range(1, 11)]
b = np.random.randint(1, 11, size=10).tolist()
```

```python
plt.plot(a, b)
plt.title("Random Values Plot (Matplotlib)")
plt.xlabel("a (1 to 10)")
plt.ylabel("b (Random values)")
```

```python
sns.lineplot(x=a, y=b, marker='o')
plt.title("Random Values Plot (Seaborn)")
plt.xlabel("a (1 to 10)")
plt.ylabel("b (Random values)")
```

## Making a dataframe and then importing the data

```python
df = pd.DataFrame({
    'col1':a,
    'col2':b 
})

sns.lineplot(x='col1', y='col2', data = df, marker='o')     # assigning columns directly to x and y
plt.xlabel("a (1 to 10)")
plt.ylabel("b (Random values)")
plt.show()  # if you dont write this code, then a label will be shown above the graph
```

```python
data = sns.load_dataset('penguins')
data
```

```python
sns.lineplot(data, x='bill_length_mm', y='flipper_length_mm')
plt.plot()
```

```python
sns.scatterplot(data=data, x='bill_length_mm', y='flipper_length_mm', hue='sex')
plt.plot()
```

```python
sns.scatterplot(data=data, x='bill_length_mm', y='flipper_length_mm', hue='sex', size='island')
plt.plot()
```

```python
sns.scatterplot(data=data, x='bill_length_mm', y='flipper_length_mm', hue='sex', size='island', style='island')
plt.plot()
```

```python
sns.scatterplot(data=data, x='bill_length_mm', y='flipper_length_mm', hue='sex', size='island', style='island', palette=['red', 'orange', 'blue', 'purple'])
plt.plot()

# 🎨 DIFFERENT COLOR FORMATS YOU CAN USE IN `palette=`
# 1️⃣  Named Seaborn categorical palettes  (good for discrete categories)
# "deep", "muted", "bright", "pastel", "dark", "colorblind"
# "tab10", "Set1", "Set2", "Set3", "Accent", "Paired", "Dark2"

# Example:
# palette="Set2"      # nice soft pastel tones for categories


# 2️⃣  Sequential palettes  (good for ordered or numeric hue values)
# "Blues", "Greens", "Oranges", "Purples", "Reds", "Greys", "YlOrBr", "BuGn", etc.

# Example:
# palette="Blues"     # light to dark blue gradient


# 3️⃣  Diverging palettes  (good when data diverges around a center)
# "coolwarm", "RdBu", "RdYlGn", "Spectral", "PiYG", "BrBG", "PuOr"

# Example:
# palette="coolwarm"  # red → white → blue gradient


# 4️⃣  Matplotlib colormaps (works for numeric hue)
# "viridis", "plasma", "magma", "inferno", "cividis", "turbo"

# Example:
# palette="viridis"   # smooth perceptually uniform color scale


# 5️⃣  Custom color lists (define your own color set)
# palette=['#FF0000', '#00FF00', '#0000FF']  # red, green, blue (hex codes)
# palette=['red', 'orange', 'blue', 'purple']  # named colors

# Example:
# sns.scatterplot(..., palette=['#e41a1c', '#377eb8', '#4daf4a'])
```

```python
sns.lineplot(data=data, x='bill_length_mm', y='flipper_length_mm', hue='sex', style='island', size='island', palette='Accent')
plt.plot()
```

```python
sns.lineplot(data=data, x='bill_length_mm', y='flipper_length_mm', hue='sex', style='island', size='island', palette='Accent', markers=['o', '>'], dashes=True)
plt.grid()
plt.title('AAAA')

plt.plot()
```
