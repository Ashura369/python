```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset('penguins')
var
```

```python
sns.scatterplot(var, x='bill_length_mm', y='bill_depth_mm', hue='sex', size='island', style='island')
plt.show()
```

```python
# using 'sizes' to fix the appropriate size
sns.scatterplot(var, x='bill_length_mm', y='bill_depth_mm', hue='island', size='sex', sizes=(20,40), style='island')
# 20 for male and 40 for female

plt.show()
```

```python
# using alpha to set the depth of colors
# the num must be btwn 0  to 1
# 0 for very light and 1 for very deep
sns.scatterplot(var, x='bill_length_mm', y='bill_depth_mm', hue='island', size='sex', sizes=(20,40), style='island', alpha=0.5)

plt.show()
```
