```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)       # only takes the columns that holds numerical values
plt.show
```

```python
sns.pairplot(var, hue='sex')       
plt.show
```

```python
# plotting only selected columns using vars

sns.pairplot(var, vars=['total_bill', 'tip'], hue='sex')       
plt.show
```

```python
# setting different variables at x axis
sns.pairplot(var, x_vars=['size', 'tip'], hue='sex')       
plt.show
```

```python
# setting different variables at y axis
sns.pairplot(var, y_vars=['size', 'tip'], hue='sex')       
plt.show
```
