## Bivariate Analysis & Multivariate Analysis

## 1 - Scatterplot (numerical - numerical) [two numerical columns]

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# to do bivariate or multivariate analysis, we will be working on 4  different datasets. Bcoz every EDA technique will not be applicable on every dataset

tips = sns.load_dataset("tips")         # inbuilt dataset in seaborn        [a restraunt data that of every customer]
flights = sns.load_dataset("flights")   # inbuilt dataset in seaborn        [US flights data of how mnany passengers travelled every month in different years]
iris = sns.load_dataset("iris")         # inbuilt dataset in seaborn        [iris dataset]
titanic = pd.read_csv("titanic.csv")    # already existing dataset in pc    [titanic data]
```

```python
# uncomment the below to see every dataset before working on them

# tips
# flights
# iris
# titanic
```

```python
print(tips.head())
# now we are going to find relationship btwn 'total_bill' and 'tip', as both being the numerical columns

# bivariate
print(f"\n\nBIVARIATE GRAPH : \n{sns.scatterplot(data=tips, x="total_bill", y="tip")} ")    # x and y hold column names of tips
# it shows in the graph that there is a linear relationship btwn the both columns, as one rises another also rises and vice-versa



```

```python
# multivariate
print(f"MULTIVARIATE GRAPH : \n{sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")}")
# hue=tips["sex"] --→ the points are colored differently for each gender (Male vs Female)
# here each point will be colored differently based on the gender categories


"""          
The below graph shows mostly males are the ones having  high bill and high tips (blue), and there are 
some outlier males and females having less bill with more tip, and some outlier females having high bill with less tips (orange).

"""
```

```python
# this is the same code but now we have added wheather the customer is a smoker or not
print(f"MULTIVARIATE GRAPH : \n{sns.scatterplot(data=tips, x="total_bill", y="tip", hue=tips["sex"], style=tips["smoker"])}")

# style in Seaborn is used to differentiate points by marker shape according to a categorical variable.

"""         
total_bill and tip are being represented on x and y axis
gender is being represented by blue and yellow color
and now wheather they are a smoker or not is being represented by x and o.

-- x-axis → total_bill
-- y-axis → tip
-- color (hue) → sex
-- marker shape (style) → smoker


"""
```

```python
# The size column in the tips dataset represents the number of people at the table for that bill
print(f"MULTIVARIATE GRAPH : \n{sns.scatterplot(data=tips, x="total_bill", y="tip", hue=tips["sex"], style=tips["smoker"], size=tips["size"])}")

"""         
total_bill and tip are being represented on x and y axis
gender is being represented by blue and yellow color
and now wheather they are a smoker or not is being represented by x and o.
Here size is also applicable on smoker and non-smoker symbol (the higher the ppl the higher the size and vice-versa)

-- x-axis → total_bill
-- y-axis → tip
-- color (hue) → sex
-- marker shape (style) → smoker
-- size of ppl on the bill

"""

```
