```python
# use_cols parameter -- used to import only the required columns

df = pd.read_csv("aug_test.csv", usecols=['enrollee_id', 'relevent_experience', 'experience', 'training_hours'])

df
```

```python
# squeeze parameter

pd.read_csv("aug_test.csv", usecols=['enrollee_id'], squeeze=True)      # it will throw error bcoz "squeeze" has been removed from recent version of pandas
```

```python
# skiprows parameter

pd.read_csv("aug_test.csv", skiprows=[0,5])            # will remove the rows idx 0 and 5
```

```python
# nrows parameter   -- used to show a specific number of rows

pd.read_csv("aug_test.csv", nrows = 10)     # will import number of rows entered here (10 rows in this case)

# this scenario is used when there is around billions of data, and its not practical to download all those datas in your computer. In such scenarios 'nrows' is used
```

```python
# on_bad_liens -- used to remove the bad lines from the code

pd.read_csv("bad_lines.csv", on_bad_lines = "skip")
```

```python
# handling dates

# by default dates are in object (string) format in pd, hence to use all the functionalities of the dates you have to convert it into the dates that the machine can recognise

pd.read_excel("IPL_2020.xlsx", parse_dates=['date']).info()
```

```python
# convertes

def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    elif name == "Mumbai Indians":
        return "MI"
    else:
        return name

pd.read_excel("IPL_2020.xlsx")

```

```python
pd.read_excel("IPL_2020.xlsx", converters={'team1' : rename})   # it will rename all the text which is "Royal Challenges Bangalore" into "RCB"
```

```python
# na_values -- to convert any value into 'NaN'

a = pd.read_csv("aug_test.csv", na_values=['Male'])
a.head(100)
```

```python
# loading a huge dataset in chunks

# currently this file holds 2129 rows and 13 cols. it will divide the data set into a chunk of 500 data set each
# this method is practical when working on a big data set bcoz downloading a bigger data on your computer and working on it might lead to breakdown of your computer

chunks = pd.read_csv("aug_test.csv", chunksize=500)

for n, i in enumerate(chunks, start = 1):
    print(f"chunk {n} : {i.shape}")
```
