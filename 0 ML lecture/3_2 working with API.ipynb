```python
url2 = "https://tourism-and-economic-impact.p.rapidapi.com/list"

payload = {}
headers2 = {
	"x-rapidapi-key": "3d72a3ed40mshf6f339ded999111p1a51e5jsn1003e2db0078",
	"x-rapidapi-host": "tourism-and-economic-impact.p.rapidapi.com",
	"Content-Type": "application/x-www-form-urlencoded"
}

response = requests.post(url2, data=payload, headers=headers2)


economy = response.json()
economy
# below we will be writing a code to convert this api key request into a pandas dataframe project
```

```python
# below we have converted 'economy' data into the pandas data frame
pd.DataFrame(economy['result']['data'])     # results --> data (inside)

```

```python
# importing only the required data and storing it in a object and making a csv file of it and uploading it on 'kaggle'

my_data = pd.DataFrame(economy['result']['data'], columns=['country', 'year', 'gdp', 'inflation', 'unemployment'])
my_data.to_csv('country and their economic dynaics.csv', index = False)
```
