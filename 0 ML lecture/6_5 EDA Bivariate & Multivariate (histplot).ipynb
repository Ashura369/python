```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")         
flights = sns.load_dataset("flights")  
iris = sns.load_dataset("iris")         
titanic = pd.read_csv("train.csv")  

titanic
```

```python
# checking how many ppl form Pclass have died and survived
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
# checking how many ppl survieved from each classes of 1, 2, and 3
pd.crosstab(titanic['Survived'], titanic['Pclass'])
```

```python
sns.heatmap(pd.crosstab(titanic['Survived'], titanic['Pclass']))    # converting the above into a heatmap
```
