```python
# lineplot is bascially used to show data based on over a period of time, like year, month, week, days, hours, minutes, etc
flights = sns.load_dataset('flights')
flights
```

```python
sns.lineplot(flights, x='year', y='passengers')
```

```python
# using pivot table to properly show the data
flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
sns.lineplot(flights.pivot_table(values='passengers', index='month', columns='year'))
```

```python
sns.lineplot(flights, x='month', y='passengers')
```

```python
sns.lineplot(flights, x='month', y='passengers', hue='year', palette='tab10', linewidth=2)
```

```python
sns.clustermap(flights.pivot_table(values='passengers', index='month', columns='year'))
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year'))
```
