# matplatlib.pylib

Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)      # when you write plot(), the numbers will be put in a graph format
plt.title("Simple Line Plot")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.show()      # if you dont write this plt.show(), the graph will not be shown, but in vs code its not case
```

Scatter Plot

```python
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]

plt.scatter(x1, y1, color='red')        # when you write scatter() the numbers will be put in a dot format
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# plt.show()
```

Bar Chart

```python
students = ['A', 'B', 'C', 'D', 'E']
marks = [85, 90, 75, 60, 95]

plt.bar(students, marks, color='red')       # the data will be presented in a bar graph format
plt.title("Student Marks")
plt.xlabel("Names")
plt.ylabel("Marks")
```

Horizontal Bar

```python
plt.barh(students, marks, color='pink')        # making horizontal bar graph
```

Histogram

```python
import random

data = [random.randint(50, 100) for i in range(100)]    # will store 100 nums btwn 50 to 100

plt.hist(data, bins=10, color='purple', edgecolor='black')
# bins=10 → divides the entire score range (50–100) into 10 equal intervals (like 50–55, 55–60, etc.)

plt.title('Histogram Example')
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
```

Pie Chart

```python
subjects = ['Math', 'Science', 'English', 'History']
marks = [80, 90, 70, 60]

plt.pie(marks, labels=subjects, autopct='%1.1f%%', startangle=90)
plt.title("Marks Distribution")

""" 
- marks → Determines the size of each slice. Bigger numbers = bigger slice.
- labels=subjects → Labels each slice with the corresponding subject name.
- autopct='%1.1f%%' → Shows the percentage value on each slice (e.g., 25.0%).
- %1.1f → one decimal place
- %% → adds the % symbol
- startangle=90 → Rotates the pie chart so the first slice starts at 90° (vertical top).
"""
```

Multiple Plots (Subplots)

```python
a = [1, 2, 3, 4, 5]
b = [2, 4, 6, 8, 10]
c = [3, 6, 9, 12, 15]

plt.subplot(1, 2, 1)        # plt.subplot(rows, columns, index)
plt.plot(a, b, color='red')
plt.title('Plot 1')

plt.subplot(1, 2, 2)        # here change the index to 1, and you will see the difference
plt.plot(a, c, color='blue')
plt.title('Plot 2')

plt.tight_layout()
# Automatically adjusts spacing between subplots
# Prevents titles or labels from overlapping
```

Box Plot

```python
data2 = [[20, 25, 30, 35, 40],
        [10, 15, 20, 25, 30],
        [5, 10, 15, 20, 25]]

plt.boxplot(data2)
plt.title("Box Plot Example")
plt.xlabel("Groups")
plt.ylabel("Values")
```

Figure Size and Save Plot

```python
d = [1, 2, 3, 4, 5]
e = [5, 10, 15, 10, 5]

plt.figure(figsize=(6, 4))      # figsize=(width[inch], height[inch])
# Useful for adjusting the plot size to make it wider or taller
plt.plot(d, e, color='orange', marker='o')      # marker='o' → marks each point with a circle

plt.title("Custom Size Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)      # prints the gridliens on the graph
plt.savefig("simple_plot.png")      # Saves the current figure as a PNG image file named simple_plot.png in the current directory


""" 
THE SYMBOLS YOU CAN USE IN PLACE OF MARKER = 'O'

'o'	Circle	
'^'	Up triangle	
'+'	Plus	
'*'	Star	
'p'	Pentagon	

's'	Square
'v'	Down triangle
'x'	X
'D'	Diamond
'h'	Hexagon

"""
```

Combining Multiple Lines
