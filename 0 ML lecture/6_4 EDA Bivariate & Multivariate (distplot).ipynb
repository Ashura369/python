# 4 - Distplot (Numerical - Categorical)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")         
flights = sns.load_dataset("flights")  
iris = sns.load_dataset("iris")         
titanic = pd.read_csv("train.csv")  
```

```python
titanic
```

```python
sns.distplot(titanic['Age'])
plt.show()
```

```python
sns.distplot(titanic['Age'], hist=False)
# hist = False will remove the bars behind the curve
plt.show()
```

```python
sns.distplot(titanic['Fare'], hist=False,color='red', label='fare')
sns.distplot(titanic['Age'], hist=False, color='green', label='age')
plt.legend()
plt.show()

# the x-axis is shared, but it represents values from both Age and Fare mixed together, not separately scaled.
```

```python
# we cant use hue in distplot so we will be using a technique where we can put multiple columns
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist=False, color='red', label='Did Not Survive')

sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist=False, color='green', label='Survived')

# add legend and title
plt.legend()        # this shows the label for each color
plt.title('Age Distribution by Survival Status')
plt.show()

```
