# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # independent variable, convered from 1D to 2D
y = np.array([2, 4, 5, 4, 5])                  # dependent variable

# Create model
model = LinearRegression()  # This line creates an empty linear regression model(as there is no values or variable that holds a value inside the paranthesis), ready to be trained.

# You are creating an object (a box) that knows how to do linear regression,
# but it doesn’t know your data yet — it hasn’t learned anything.
# It’s like a blank student who knows how to learn, but hasn’t studied yet.


# Train model
model.fit(X, y)

# The model calculates and fills in the missing parts:
    # finds the best m (slope)
    # finds the best c (intercept)



# Predict
y_pred = model.predict(X)

# Display
print("Coefficient (m):", model.coef_)
print("Intercept (c):", model.intercept_)


"""
Create an empty model	    -        model = LinearRegression()	   -     You hire a teacher who knows how to find straight lines.
Train the model	            -        model.fit(X, y)	           -     You give the teacher your data — they study and find the best-fit line.
Predict                     -    	model.predict(X)	           -     The teacher now uses what they learned to give you predictions.

"""


# Plot
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Best Fit Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
