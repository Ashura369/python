# Pairplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")         
flights = sns.load_dataset("flights")  
iris = sns.load_dataset("iris")         
titanic = pd.read_csv("train.csv")  

iris
```

```python
# its bascially a collection of scatterplot
# instead of plotting each and every column as scatterplot, you can use pairplot to save time
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```
