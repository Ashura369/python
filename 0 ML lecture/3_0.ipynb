Extracting input and output col

```python
# storing the independent and dependent data sets into different variables
a = df.iloc[:, 0:2]
b = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

train_test_split(a, b, test_size=0.1)
# first variable a represents x axis, and the second variable b represents y axis
# when you write test_size = 0.1 that means out all the datas 10 percent of data to be used for 'testing' and rest of the data should be used for 'training'
# here it will give 4 outputs training and test data of a and training and test data of b, in total it will give 4 data sets you can store them into four different variables
# remember the order it will give data first it will give train and test data of a and then it will give train and test data of b


# storing the data into four different variables
x_train, x_test, y_train, y_test = train_test_split(a, b, test_size=0.1)

```
